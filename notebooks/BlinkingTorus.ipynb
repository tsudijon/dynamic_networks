## Blinking Torus example
How does this work?

```python
np.random.seed(10)
K = 8
NPeriods = 10
TransLen = 5
res = gpg.getBlinkingVideo(K, NPeriods, TransLen)
[XRet, A, idx2pos] = [res['XRet'], res['A'], res['idx2pos']]

gpg.draw2DGraph(A, idx2pos, np.zeros(A.shape[0]), True)
plt.show()
```

```python
#First plot that scalar function over time at each individual node
plt.figure(figsize=(10, XRet.shape[1]))
for j in range(XRet.shape[1]):
    plt.subplot(XRet.shape[1], 1, j+1)
    plt.plot(XRet[:, j])
    plt.ylim([-0.2, 1.2])
    if j == XRet.shape[1]-1:
        ax = plt.gca()
        ax.set_yticks([])
    else:
        plt.axis('off')
    plt.title("%i"%j)
plt.tight_layout()
plt.show()

```

## Apply pipeline
